## Motivation

Rossmann là một trong những chuỗi siêu thị lớn nhất ở Đức, với hơn 3.000 cửa hàng trên toàn quốc và doanh thu hàng năm khoảng 10 tỷ Euro. Dự đoán doanh số bán hàng chính xác là rất quan trọng cho việc quản lý hiệu quả các hoạt động kinh doanh và duy trì lợi thế cạnh tranh.

### Tại sao cần dự đoán doanh số bán hàng?

1. **Quản lý hàng tồn kho (Inventory Management)**:
   - Tránh thiếu hàng: Giảm mất doanh thu từ khách hàng không tìm thấy sản phẩm
   - Tránh tồn kho quá nhiều: Giảm chi phí lưu trữ, hư hỏng, và lãng phí
   - Ước tính: Sai sót trong dự đoán có thể làm tăng chi phí lên đến 20-30%

2. **Lập kế hoạch nhân sự (Staff Planning)**:
   - Sắp xếp nhân viên phù hợp với lưu lượng khách hàng
   - Tránh chi phí nhân sự thừa hoặc thiếu
   - Tác động: Nhân sự chiếm 10-15% chi phí vận hành

3. **Chiến lược marketing và khuyến mãi (Marketing Strategies)**:
   - Đo lường hiệu quả của các chương trình khuyến mãi
   - Tối ưu hóa thời điểm và quy mô khuyến mãi
   - Ví dụ: Promo có thể tăng doanh số 10-50% tùy loại

4. **Ra quyết định tài chính (Financial Planning)**:
   - Dự báo doanh thu chính xác cho lập ngân sách
   - Quản lý dòng tiền và đầu tư
   - Giảm rủi ro tài chính từ biến động doanh số

5. **Cạnh tranh với đối thủ (Competitive Advantage)**:
   - Sử dụng dữ liệu để đưa ra quyết định nhanh hơn đối thủ
   - Tối ưu hóa giá cả và sản phẩm dựa trên dự đoán nhu cầu
   - Trong ngành bán lẻ, lợi thế cạnh tranh có thể mang lại tăng trưởng 5-10% doanh thu

### Ngữ cảnh bài toán

- **Dữ liệu lịch sử**: Từ 2013-01-01 đến 2015-07-31 cho 1.115 cửa hàng
- **Dự đoán**: Doanh số cho 6 tuần tiếp theo (từ 2015-08-01 đến 2015-09-17)
- **Yếu tố ảnh hưởng**: 
  - Thời gian: Ngày trong tuần, tháng, năm, ngày lễ
  - Khuyến mãi: Promo hàng ngày và Promo2 liên tục
  - Cạnh tranh: Khoảng cách và thời điểm mở cửa đối thủ
  - Cửa hàng: Loại cửa hàng, mức độ assortment
- **Thách thức**: Biến động theo mùa, tác động của ngày lễ, hiệu ứng khuyến mãi

### Tác động kinh doanh của dự đoán chính xác

- **Giảm chi phí**: Tối ưu inventory và nhân sự có thể tiết kiệm 5-15% chi phí
- **Tăng doanh thu**: Dự đoán tốt hơn giúp tăng hiệu quả marketing và giảm mất bán
- **Cải thiện trải nghiệm khách hàng**: Đảm bảo sản phẩm có sẵn khi cần
- **Lợi nhuận tổng thể**: Mô hình dự đoán tốt có thể cải thiện lợi nhuận 2-5%

## Evaluation Metrics

Trong bài toán dự đoán doanh số bán hàng, chúng ta cần các metrics để đánh giá độ chính xác của mô hình. Với bộ dữ liệu Rossmann, metric chính được sử dụng là **Root Mean Square Percentage Error (RMSPE)**.

### Tại sao dùng RMSPE?

- **Percentage Error**: Đo lường sai số theo tỷ lệ phần trăm, phù hợp với doanh số có thể thay đổi lớn giữa các cửa hàng.
- **Root Mean Square**: Phạt nặng các lỗi lớn hơn, khuyến khích mô hình tránh dự đoán sai quá nhiều.
- **Normalized**: Giúp so sánh giữa các cửa hàng có quy mô khác nhau.

### Công thức RMSPE

$\text{RMSPE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} \left( \frac{y_i - \hat{y}_i}{y_i} \right)^2}$

Trong đó:
- $n$: Số lượng mẫu (cửa hàng × ngày)
- $y_i$: Giá trị thực tế của doanh số (Sales)
- $\hat{y}_i$: Giá trị dự đoán
- Lưu ý: Chỉ tính cho các cửa hàng mở (Open = 1) và Sales > 0

### Ý nghĩa

- **Giá trị thấp tốt hơn**: RMSPE = 0 nghĩa là dự đoán hoàn hảo.
- **Ví dụ**: Nếu RMSPE = 0.1 (10%), nghĩa là trung bình sai số dự đoán là 10% so với doanh số thực tế.
- **Trong Kaggle**: Public leaderboard dùng 30% dữ liệu test, private dùng 70% còn lại.

### Các metrics khác có thể dùng

1. **Mean Absolute Error (MAE)**:
   $\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$
   - Ý nghĩa: Sai số trung bình tuyệt đối, dễ hiểu.

2. **Root Mean Square Error (RMSE)**:
   $\text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$
   - Ý nghĩa: Sai số trung bình bình phương, phạt lỗi lớn.

3. **Mean Absolute Percentage Error (MAPE)**:
   $\text{MAPE} = \frac{100}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right|$
   - Ý nghĩa: Tỷ lệ sai số trung bình theo phần trăm.

4. **R-squared (R²)**:
   $R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$
   - Ý nghĩa: Tỷ lệ biến thiên được giải thích bởi mô hình (0-1, cao tốt hơn).

````python
import numpy as np

# Ví dụ tính RMSPE
def rmspe(y_true, y_pred):
    """
    Root Mean Square Percentage Error
    Chỉ tính cho y_true > 0
    """
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    mask = y_true > 0
    y_true = y_true[mask]
    y_pred = y_pred[mask]
    return np.sqrt(np.mean(((y_true - y_pred) / y_true) ** 2))

# Ví dụ dữ liệu
y_true = [100, 200, 150, 300, 250]
y_pred = [90, 210, 160, 280, 260]

rmspe_value = rmspe(y_true, y_pred)
print(f"RMSPE: {rmspe_value:.4f} ({rmspe_value*100:.2f}%)")

# So sánh với RMSE và MAE
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((np.array(y_true) - np.array(y_pred)) ** 2))

def mae(y_true, y_pred):
    return np.mean(np.abs(np.array(y_true) - np.array(y_pred)))

rmse_value = rmse(y_true, y_pred)
mae_value = mae(y_true, y_pred)

print(f"RMSE: {rmse_value:.2f}")
print(f"MAE: {mae_value:.2f}")
````

### Ví dụ tính toán Metrics

Với dữ liệu mẫu:
- y_true = [100, 200, 150, 300, 250]
- y_pred = [90, 210, 160, 280, 260]

Kết quả:
- **RMSPE**: 0.0678 (6.78%) - Sai số trung bình 6.78% theo tỷ lệ
- **RMSE**: 12.65 - Sai số trung bình bình phương (đơn vị tiền tệ)
- **MAE**: 12.00 - Sai số trung bình tuyệt đối (đơn vị tiền tệ)

**Nhận xét**:
- RMSPE thấp (6.78%) cho thấy dự đoán khá chính xác
- RMSE và MAE cho thấy sai số tuyệt đối khoảng 12-13 đơn vị
- Trong thực tế với Rossmann, RMSPE mục tiêu thường dưới 15%

## Giới thiệu chi tiết về DVC (Data Version Control)

### DVC là gì?

DVC (Data Version Control) là một công cụ mã nguồn mở được phát triển bởi Iterative.ai, giúp quản lý phiên bản dữ liệu và machine learning models. DVC giải quyết vấn đề mà Git không thể xử lý tốt: quản lý files lớn và dữ liệu nhị phân.

**Sự khác biệt với Git:**
- Git: Version control cho code (text files)
- DVC: Version control cho data và models (binary files)
- DVC sử dụng Git để lưu metadata, còn data thực tế lưu ở local/remote storage

### Tại sao cần DVC trong ML projects?

1. **Dữ liệu lớn**: Datasets thường GB/TB, không thể commit trực tiếp vào Git
2. **Reproducibility**: Đảm bảo experiments có thể chạy lại với cùng data
3. **Collaboration**: Team members có thể share datasets mà không duplicate
4. **Experiment tracking**: Theo dõi thay đổi của models và metrics
5. **Deployment**: Version control cho production models

### Cài đặt và khởi tạo DVC

```bash
# Cài đặt DVC
pip install dvc

# Khởi tạo trong project (tạo .dvc folder)
dvc init

# Kiểm tra
ls -la
# Sẽ thấy .dvc/ folder
```

### Workflow cơ bản

#### 1. Thêm dữ liệu vào tracking
```bash

dvc add data/train.csv

# Kết quả: tạo train.csv.dvc (metadata file)
# train.csv được move vào .dvc/cache/
```

#### 2. Commit metadata vào Git
```bash
git add train.csv.dvc .gitignore
git commit -m "Add training data"
```

#### 3. Push data lên remote storage
```bash
# Cấu hình remote (ví dụ S3)
dvc remote add -d myremote s3://mybucket/path

# Push data
dvc push
```

#### 4. Clone và pull data
```bash
# Clone repo
git clone <repo>
cd <repo>

# Pull data từ remote
dvc pull
```

### Giải thích cụ thể về .dvcignore

#### .dvcignore là gì?

`.dvcignore` là file cấu hình trong DVC, tương tự như `.gitignore` trong Git. Nó định nghĩa các patterns của files/folders mà DVC sẽ **không track** (ignore) khi thực hiện các lệnh như `dvc add`, `dvc status`, hoặc `dvc diff`.

#### Tại sao cần .dvcignore?

1. **Performance**: Loại bỏ files không cần thiết khỏi scanning, tăng tốc độ
2. **Security**: Tránh track files nhạy cảm (passwords, keys)
3. **Clean tracking**: Chỉ track data thực sự cần thiết cho ML pipeline
4. **Storage optimization**: Giảm size của DVC cache và remote storage

#### Syntax và Patterns

`.dvcignore` sử dụng glob patterns tương tự `.gitignore`:

```gitignore
# Comments bắt đầu bằng #
*.tmp                    # Ignore tất cả files .tmp
logs/                    # Ignore folder logs/
!important.log          # Exception: track important.log
models/*.pkl            # Ignore model files .pkl
data/raw/*              # Ignore files trong data/raw/
**/temp/                # Ignore tất cả temp/ folders ở mọi level
```

#### Ví dụ thực tế cho Rossmann project

```gitignore
# Temporary files
*.tmp
*.temp
*~

# Logs
logs/
*.log

# Cache files
__pycache__/
*.pyc

# Jupyter notebook checkpoints
.ipynb_checkpoints/

# Model artifacts (nếu không muốn track)
models/checkpoints/

# Raw data (nếu đã processed)
data/raw/

# IDE files
.vscode/

# OS files
.DS_Store
Thumbs.db

# DVC internal
.dvc/tmp/
```

#### Sự khác biệt với .gitignore

| Aspect | .gitignore | .dvcignore |
|--------|------------|------------|
| Mục đích | Ignore files khỏi Git | Ignore files khỏi DVC tracking |
| Áp dụng | Git commands (add, status) | DVC commands (add, status, diff) |
| Files track | Code, config, docs | Data, models, large files |
| Khi nào check | Trước Git operations | Trước DVC operations |

#### Cách sử dụng

1. **Tạo file .dvcignore**:
   ```bash
   touch .dvcignore
   ```

2. **Thêm patterns**:
   ```bash
   echo "*.log" >> .dvcignore
   echo "temp/" >> .dvcignore
   ```

3. **Kiểm tra**:
   ```bash
   dvc status  # Sẽ không show ignored files
   ```

#### Best practices

1. **Ignore early**: Thêm patterns trước khi `dvc add`
2. **Be specific**: Sử dụng patterns chính xác để tránh ignore nhầm
3. **Document**: Comment giải thích tại sao ignore
4. **Review regularly**: Update khi project structure thay đổi
5. **Team consistency**: Share .dvcignore trong team

#### Lưu ý quan trọng

- `.dvcignore` chỉ ảnh hưởng đến DVC operations, không ảnh hưởng Git
- Files đã được track bởi DVC sẽ không bị ignore (dùng `dvc remove` để untrack)
- Patterns được evaluate từ root của repository
- DVC sẽ vẫn scan ignored files để check, nhưng không track chúng

Trong project Rossmann, .dvcignore giúp loại bỏ temporary files, logs, và cache để chỉ track datasets và models thực sự cần thiết.
